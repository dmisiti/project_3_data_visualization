```python
import matplotlib.pyplot as plt

# Filter the DataFrame for rows where the district is 'Connecticut'
connecticut_df = race_ethnicity_df[race_ethnicity_df['District name'] == 'Connecticut']

# Group by 'Ethnicity' and calculate the average attendance rates for each year
attendance_rates = connecticut_df.groupby('Student group')[['2019-2020 attendance rate', '2020-2021 attendance rate', '2021-2022 attendance rate - year to date']].mean()

# Plotting the data
attendance_rates.plot(kind='bar', figsize=(12, 7), color=['skyblue', 'lightgreen', 'salmon'])

# Adding labels on top of each bar
ax = plt.gca()
for i in range(len(attendance_rates)):
    for j, year in enumerate(attendance_rates.columns):
        ax.text(i - 0.2 + j * 0.2, attendance_rates.iloc[i, j] + 0.001, f'{attendance_rates.iloc[i, j]:.2f}', ha='center', va='bottom', fontsize=10)

# Add a horizontal line at the national average (0.9 attendance rate)
plt.axhline(y=0.9, color='red', linestyle='--', label='National Average (0.9)')

# Title and labels
plt.title('Average Attendance Rates by Ethnicity in Connecticut Districts (2019-2022)')
plt.xlabel('Ethnicity')
plt.ylabel('Average Attendance Rate')
plt.xticks(rotation=45)

# Add the legend with a smaller font size and a better position
plt.legend(title='Year', labels=['National Average', '2019-2020', '2020-2021', '2021-2022'], loc='upper left', fontsize=8, bbox_to_anchor=(1, 1))

plt.show()


```

```python
import matplotlib.pyplot as plt

# Filter the DataFrame for rows where the district is 'Connecticut'
connecticut_df = race_ethnicity_df[race_ethnicity_df['District name'] == 'Connecticut']

# Group by 'Ethnicity' and calculate the average attendance rates for each year
attendance_rates = connecticut_df.groupby('Student group')[['2019-2020 attendance rate', '2020-2021 attendance rate', '2021-2022 attendance rate - year to date']].mean()

# Transpose the DataFrame to have years as the x-axis and ethnicities as bars
attendance_rates = attendance_rates.transpose()

# Generate a unique color for each ethnicity
colors = plt.cm.get_cmap('tab20', len(attendance_rates.columns))  # Use a colormap with enough colors

# Plotting the data with a different color for each ethnicity
attendance_rates.plot(kind='bar', figsize=(12, 7), color=colors(range(len(attendance_rates.columns))))

# Adding labels at the top of each bar
ax = plt.gca()
for i in range(len(attendance_rates)):
    for j, ethnicity in enumerate(attendance_rates.columns):
        # Adjust the label position to sit closely at the top of the bar
        ax.text(i - 0.2 + j * 0.13, attendance_rates.iloc[i, j] + 0.001, f'{attendance_rates.iloc[i, j]:.2f}', 
                ha='center', va='bottom', fontsize=10)

# Add a horizontal line at the national average (0.9 attendance rate)
plt.axhline(y=0.9, color='red', linestyle='--', label='National Average (0.9)')

# Title and labels
plt.title('Average Attendance Rates by Year in Connecticut Districts (2019-2022)')
plt.xlabel('Year')
plt.ylabel('Average Attendance Rate')
plt.xticks(rotation=0)

# Adjust the y-axis to start at 0.7
plt.ylim(0.7, 1.0)  # Set the lower limit to 0.7, and upper limit to 1.0 for better scaling

# Add the legend with a smaller font size and a better position
plt.legend(title='Ethnicity', loc='upper left', fontsize=8, bbox_to_anchor=(1, 1))

plt.show()



```
