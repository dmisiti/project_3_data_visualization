```python
# Dependencies
from sqlalchemy import create_engine
import pandas as pd
```

```python
# Create engine
engine = create_engine('postgresql://postgres:postgres@localhost:5432/attendance_db')
```

```python
# Read in Attendance table
attendance_df = pd.read_sql('SELECT * from "Attendance"', engine)
attendance_df.head()
```

```python
# Read in Locations table
locations_df = pd.read_sql('SELECT * from "Locations"', engine)
locations_df.head()
```

```python
# Save tables as CSV files
attendance_df.to_csv('resources/attendance_table.csv', index=False)
locations_df.to_csv('resources/locations_table.csv', index=False)
```
