```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in CSV
attendance_df = pd.read_csv("resources/attendance_and_locations.csv")
```

```python
# Display the data
attendance_df.head()
```

```python
attendance_columns = [
    '2019-2020 attendance rate', 
    '2020-2021 attendance rate', 
    '2021-2022 attendance rate - year to date'
]

# Convert rates into percentages
attendance_df[attendance_columns] = attendance_df[attendance_columns] * 100
attendance_df.head()
```

# Line Graph 1: Overall Attendance in Connecticut

```python
# Filter the data for all of Connecticut
connecticut_data = attendance_df[attendance_df['District name'] == 'Connecticut']
connecticut_data
```

```python
drop_columns = [
    'Category', 
    '2021-2022 student count - year to date', 
    '2020-2021 student count', 
    '2019-2020 student count', 
    'Latitude', 
    'Longitude'
]

connecticut_data = connecticut_data.drop(columns=drop_columns)
connecticut_data
```

```python
# Filter the data for the three groups: All Students, Students With High Needs, and Students Without High Needs
all_students_data = connecticut_data[connecticut_data['Student group'] == 'All Students']
high_needs_data = connecticut_data[connecticut_data['Student group'] == 'Students With High Needs']
low_needs_data = connecticut_data[connecticut_data['Student group'] == 'Students Without High Needs']
```

```python
# Extract the years and attendance rates for each group
years = ['2019-2020', '2020-2021', '2021-2022']

attendance_all_students = all_students_data[attendance_columns].values.flatten()
attendance_high_needs = high_needs_data[attendance_columns].values.flatten()
attendance_low_needs = low_needs_data[attendance_columns].values.flatten()
```

```python
# Create the plot
plt.figure(figsize=(10, 6))

# Plot each group's data
plt.plot(years, attendance_all_students, marker='o', label='All Students', color='black', linewidth = 1)
plt.plot(years, attendance_high_needs, marker='o', label='Students With High Needs', color='b', linewidth = 3)
plt.plot(years, attendance_low_needs, marker='o', label='Students Without High Needs', color='g', linewidth = 3)

# Add a vertical line to mark the start of the COVID pandemic
plt.axvline(x=0.5, color='black', linestyle='--')

# Add shaded regions for Pre-pandemic and Post-pandemic
plt.axvspan(0, 0.5, color='skyblue', alpha=0.2, label='Pre-pandemic')  # Shaded area for pre-pandemic
plt.axvspan(0.5, 2, color='lightcoral', alpha=0.2, label='Post-pandemic')  # Shaded area for post-pandemic

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Attendance Rate (%)')
plt.title('Attendance Rates in Connecticut by Student Needs')

# Add a legend
plt.legend()

# Show the plot
plt.grid(False)
plt.show()
```
